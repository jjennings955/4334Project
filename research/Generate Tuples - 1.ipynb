This iPython notebook searches for tuples that contain likely foods using a regular expression.
The idea is that a very common pattern for menu items is "the Word Word Word" The regular expression:

the [A-Z][a-z]+ .*?

captures this information.


```python
__author__ = 'jason'
import pandas as pd
import re

d = pd.read_csv(r'./data/sorted_out_2.csv', delimiter='\t', error_bad_lines=False)
d.columns=[u'filename', u'sentence number', u'arg1', u'rel', u'arg2', u'arg1 start', u'arg1 end', u'rel start', u'rel end', u'arg2 start', u'arg2 end', u'conf', u'words', u'tags', u'chunks', u'a1norm', u'relnorm', u'a2norm']
```

```python
pattern = re.compile('the [A-Z][a-z]+ .*?')
possible_foods = set([i for i in d['arg2'].values if pattern.match(str(i))])

def matches_pattern(s):
    try:
        return pattern.match(s) != None
    except Exception:
        return False
```

```python
condition1 = d['arg2'].apply(matches_pattern)
condition2 = d['arg1'].apply(matches_pattern)
```

```python
s1 = d[condition2]
s2 = d[condition1]
```

```python
import csv
s1.to_csv('type1.csv', delimiter=',', quoting=csv.QUOTE_ALL)
s2.to_csv('type2.csv', delimiter=',', quoting=csv.QUOTE_ALL)
```
