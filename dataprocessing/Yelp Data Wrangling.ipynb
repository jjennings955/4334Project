```python
import pandas as pd
import csv

def shuffle(df):
    """
    A function to shuffle a pandas dataframe
    Source: http://stackoverflow.com/questions/15772009/shuffling-permutating-a-dataframe-in-pandas
    JeromeZhao
    """
    import random
    index = list(df.index)
    random.shuffle(index)
    df = df.ix[index]
    df.reset_index()
    return df
```

Join Restaurants.csv and Reviews.csv on business_id
===================================================
This will give us a dataframe with reviews pertaining only to restaurants

```python
businesses = pd.read_csv('restaurants.csv', encoding='cp1252')
reviews = pd.read_csv('reviews.csv', encoding='cp1252')
```

```python
restaurant_reviews = pd.merge(reviews, businesses, on='business_id', how='inner')
```

Break the File into Manageable Chunks
=====================================
The dataset is very large. Memory expensive operations on the complete dataset can exceed most machine's maximum memory.

```python
shuffled_reviews = shuffle(restaurant_reviews)
for i in range(15):
    print("Creating chunk %d" % (i))
    sampled_reviews = shuffled_reviews[i*50000:(i+1)*50000]
    sampled_reviews['text'] = sampled_reviews['text'].str.replace('\n', '').replace('\r', '')
    print(sampled_reviews)
    print("Writing chunk %d to disk" % (i))
    sampled_reviews.to_csv('sampled-reviews-test-%d.csv' % (i), sep='|', header=True, columns=['review_id', 'text'], encoding='cp1252', quoting=csv.QUOTE_ALL)
```
